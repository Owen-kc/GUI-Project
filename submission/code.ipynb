## Data Pre-Processing
***
 We then must import the relevant libraries, and define the file paths for each of our csv files. Then, we can load the datasets, combine them, and print the result.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define file paths 
opel_corsa_01_path = './opel_corsa_01.csv'
opel_corsa_02_path = './opel_corsa_02.csv'
peugeot_207_01_path = './peugeot_207_01.csv'
peugeot_207_02_path = './peugeot_207_02.csv'

# Load the datasets
opel_corsa_01_df = pd.read_csv(opel_corsa_01_path, sep=';', decimal=',')
opel_corsa_02_df = pd.read_csv(opel_corsa_02_path, sep=';', decimal=',')
peugeot_207_01_df = pd.read_csv(peugeot_207_01_path, sep=';', decimal=',')
peugeot_207_02_df = pd.read_csv(peugeot_207_02_path, sep=';', decimal=',')

# Combine the datasets into one
combined_df = pd.concat([opel_corsa_01_df, opel_corsa_02_df, peugeot_207_01_df, peugeot_207_02_df], ignore_index=True)

# Print the combined dataset, and the first few rows with head
print (combined_df)

combined_df.head()
```

Next, we will continue with pre-processing the data by ensuring that missing values are dropped, and any unnamed columns are removed

```python
# Drop rows with any missing value
combined_df.dropna(inplace=True)

# Drop the unnamed column 
if 'Unnamed: 0' in combined_df.columns:
    combined_df.drop(columns=['Unnamed: 0'], inplace=True)

# Drop the variables we dont need
combined_df = combined_df.drop(['roadSurface', 'traffic'], axis=1)

# Check if there are any missing values
print("Number of missing values after dropping NA:", combined_df.isnull().sum().sum())

# exploration of the target variable distribution
print(combined_df['drivingStyle'].value_counts())
```

We can see that the datatypes are in an object state, we should convert the relevant numerical datatypes for our models to process better. We will also fill NaN values and check the data types.

```python
# Convert numerical features from object to float64
numerical_columns = ['AltitudeVariation', 'VehicleSpeedInstantaneous', 'VehicleSpeedAverage',
                     'VehicleSpeedVariance', 'VehicleSpeedVariation', 'LongitudinalAcceleration',
                     'EngineLoad', 'EngineCoolantTemperature', 'ManifoldAbsolutePressure',
                     'EngineRPM', 'MassAirFlow', 'IntakeAirTemperature', 'VerticalAcceleration',
                     'FuelConsumptionAverage']

for col in numerical_columns:
    combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')  

# Forward fill NaN values
combined_df.fillna(method='ffill', inplace=True) 

# Re-check data types after conversion
print("\nUpdated Data types after conversion:")
print(combined_df.dtypes)
```

In the pre-processing step, we have done the following:

- Loaded and combined multiple CSV files into a single DataFrame.
- Removed rows with missing values.
- Removed unnessecary columns.
- Converted relevant numeric columns to float64 for model compatibility.
- Applied forward fill to handle remaining NaN values.

Now that we can see the extent of the imbalance, we can generate a more detailed classification report. The direct calculation of metrics above calculates the metrics in a more simple, concise way, without considering class imbalance. Now that we know there is a class imbalance, we must identify where the models struggle and how we can deal with that.

```python
from sklearn.metrics import classification_report

# For KNN
print("\nk-Nearest Neighbors Metrics:")
print(classification_report(y_test, y_test_pred_knn, target_names=['AggressiveStyle', 'EvenPaceStyle']))

# For SVC
print("\nSupport Vector Classifier Metrics:")
print(classification_report(y_test, y_test_pred_svc, target_names=['AggressiveStyle', 'EvenPaceStyle']))   

# For Logistic Regression
print("\nLogistic Regression Testing Metrics:")
print(classification_report(y_test, y_test_pred, target_names=['AggressiveStyle', 'EvenPaceStyle']))
```

This data shows us the imbalance. The support values are disproportionate between negative and positive (kNN - 828/6301), which indicates that the models performance is biased towards the majority class (EvenPaceStyle / 1). The recall rate is relatively low, which tells us that the models struggle to identify instances of the minority class. This means that our model performs well on classifying the majority class but poorly on the minority class. To address this concern, we can use data balancing techniques and compare the results. While Logistic Regression and SVM provide the `class_weight` parameter for data balancing, kNN lacks this option, and if we used `class_weight` for two models and SMOTE for another, the results might be biased. Therefore, it's best to utilize the Synthetic Minority Over-sampling Technique (SMOTE) to address the imbalance specifically for kNN. This technique addresses imbalanced datasets by generating synthetic samples for the minority class.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE

# Define a function to train and evaluate a model with SMOTE
def train_evaluate_model(X_train, y_train, X_test, y_test, model_name, model):
    smote = SMOTE()
    X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
    
    model.fit(X_train_smote, y_train_smote)
    y_test_pred = model.predict(X_test)
    
    print(f"\n{model_name} Metrics:")
    print("Accuracy:", accuracy_score(y_test, y_test_pred))
    print("Precision:", precision_score(y_test, y_test_pred))
    print("Recall:", recall_score(y_test, y_test_pred))
    print("F1 Score:", f1_score(y_test, y_test_pred))
    
    return y_test, y_test_pred

# Train and evaluate Logistic Regression with SMOTE
y_test_log_reg, y_test_pred_log_reg = train_evaluate_model(X_train, y_train, X_test, y_test, "Logistic Regression with SMOTE", log_reg)

# Train and evaluate Support Vector Classifier with SMOTE
y_test_svc, y_test_pred_svc = train_evaluate_model(X_train, y_train, X_test, y_test, "Support Vector Classifier with SMOTE", svc)

# Train and evaluate k-Nearest Neighbors with SMOTE
y_test_knn, y_test_pred_knn = train_evaluate_model(X_train, y_train, X_test, y_test, "k-Nearest Neighbors with SMOTE", knn)
```

Again, we get the metrics above without the consideration for the class imbalance. With these new metrics, we can now analyze the classification report and understand if our models now handle the unbalanced side of the data more efficiently. 

```python
from sklearn.metrics import classification_report

# Generate and print the classification report for k-Nearest Neighbors with SMOTE
print("\nk-Nearest Neighbors Metrics:")
print(classification_report(y_test_knn, y_test_pred_knn, target_names=['AggressiveStyle', 'EvenPaceStyle']))

# Generate and print the classification report for Support Vector Classifier with SMOTE
print("\nSupport Vector Machine Metrics:")
print(classification_report(y_test_svc, y_test_pred_svc, target_names=['AggressiveStyle', 'EvenPaceStyle']))

# Generate and print the classification report for Logistic Regression with SMOTE
print("\nLogistic Regression Testing Metrics:")
print(classification_report(y_test_log_reg, y_test_pred_log_reg, target_names=['AggressiveStyle', 'EvenPaceStyle']))
```

The goal of applying SMOTE was to increase the model's recognition and handling of the minority class. The main attribute that identify the models ability to deal with the minority class recognition is recall, which represents the true positive rate. Before applying SMOTE, the recall rate was relatively low for all the models. 0.57 for kNN, 0.11 for SVC and 0.00 for Logistic Regression. After SMOTE, kNN incerased to 0.87, SVM increased to 0.84 and Logistic Regression increased to 0.75.

## Cross Validation
***

Now, we should cross validate to ensure we have the most consistent results across the board.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
import numpy as np


# Initialize models
log_reg = LogisticRegression(max_iter=1000)
svc = SVC()
knn = KNeighborsClassifier(n_neighbors=6)

# Create SMOTE pipelines as before
log_reg_pipe = Pipeline([('smote', SMOTE(random_state=42)), ('model', log_reg)])
svc_pipe = Pipeline([('smote', SMOTE(random_state=42)), ('model', svc)])            # retrain w smote
knn_pipe = Pipeline([('smote', SMOTE(random_state=42)), ('model', knn)])

# Define models and names for iteration
models = [log_reg_pipe, svc_pipe, knn_pipe]
names = ['Logistic Regression', 'SVC', 'KNN']

# Perform cross-validation
for model, name in zip(models, names):
    # cv=3 for quicker runtime, n_jobs=-1 to use all processors
    scores = cross_val_score(model, X, y, cv=3, scoring='accuracy', n_jobs=-1)
    print(f"{name} Accuracy: {np.mean(scores):.4f} (+/- {np.std(scores)*2:.4f})")
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer
from imblearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE

# Define models
log_reg = LogisticRegression(max_iter=3000) # set to 3000 to ensure convergence
svc = SVC()
knn = KNeighborsClassifier(n_neighbors=6)

# Define scoring metrics (from classification_report)
scoring = {'accuracy': 'accuracy',
           'precision': 'precision',
           'recall': 'recall',
           'f1_score': 'f1'}

# Create pipelines with SMOTE
log_reg_pipe = Pipeline([('smote', SMOTE(random_state=42)), ('model', log_reg)])
svc_pipe = Pipeline([('smote', SMOTE(random_state=42)), ('model', svc)])
knn_pipe = Pipeline([('smote', SMOTE(random_state=42)), ('model', knn)])

# cross-validation
models = {'Logistic Regression': log_reg_pipe,
          'Support Vector Machine': svc_pipe,
          'k-Nearest Neighbors': knn_pipe}

for name, pipeline in models.items():
    cv_results = cross_validate(pipeline, X, y, cv=5, scoring=scoring)
    print(f"\n{name} - Cross-validation Metrics:")
    for metric in scoring:
        print(f"{metric.capitalize()}: {cv_results['test_' + metric].mean()}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_validate

# Collecting results
results = []

for name, pipeline in models.items():
    cv_results = cross_validate(pipeline, X, y, cv=5, scoring=scoring)
    for metric in scoring:
        results.append({
            'Model': name,
            'Metric': metric.capitalize(),
            'Score': cv_results['test_' + metric].mean() * 100  # mult by 100 to get percentage
        })

# Convert results into a DataFrame
results_df = pd.DataFrame(results)

# Pivot the DataFrame to format it for plotting
results_pivot = results_df.pivot(index='Metric', columns='Model', values='Score')

# Plotting
ax = results_pivot.plot(kind='bar', figsize=(8, 6), width=0.8)  # Thicker bars
plt.title('Cross-validation Metrics Comparison (SMOTE)')
plt.ylabel('Score (%)')  # Update label
plt.xticks(rotation=45)
plt.legend(title='Model', loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=len(models.keys()))  # Use the actual number of models
plt.tight_layout()

for p in ax.patches:
    ax.annotate(f"{int(round(p.get_height()))}%",  # Round to nearest whole number and add percent sign
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5),  # 5 points vertical offset
                textcoords='offset points')

plt.show()
```
